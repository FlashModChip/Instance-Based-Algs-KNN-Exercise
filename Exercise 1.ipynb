```python
# Importing libraries
import pandas as pd
import numpy as np
```

Laden des CSV und sicherung der min max Values für die Normalisierung

```python
ds = pd.read_csv('Kleider.csv',';')
```

```python
min_koerper, max_koerper = ds.min(axis=0)[0], ds.max(axis=0)[0]
min_gewicht, max_gewicht = ds.min(axis=0)[1], ds.max(axis=0)[1]
```

```python
ds
```

```python
new_ds = ds.copy()
```

Normalisierung. Die Werte ergeben 0 bis 1

```python
for y in range(2):
    for x in range(18):
        new_ds.iloc[x,y] = (ds.iloc[x,y]-ds.min(axis=0)[y])/(ds.max(axis=0)[y]-ds.min(axis=0)[y])
        # print('ds: ', x, y, ds.iloc[x,y])
        # print('min y: ', y, ds.min(axis=0)[y])
        # print((ds.iloc[x,y]-ds.min(axis=0)[y])/(ds.max(axis=0)[y]-ds.min(axis=0)[y]))
```

```python
new_ds
```

```python
predict_ds = new_ds.copy()
```

Euklid Berechnung mit übergabe der zu bestimmenden Werten

```python
def euklid (height, weight):
    norm_h = ((height-min_koerper)/(max_koerper-min_koerper))
    norm_w = (weight-min_gewicht)/(max_gewicht-min_gewicht)
    arr=[]
    
    for x in range(18):
        arr.append(np.square((norm_h-new_ds.iloc[x,0])**2 + (norm_w-new_ds.iloc[x,1])**2))
    return arr
```

```python
distances = []
distances = euklid(163, 61)
```

```python
distances
```

Erweiterung der Tabelle um die Distanz

```python
predict_ds.insert(3, "distanz", distances, True)
```

```python
predict_ds
```

Sortierung nach der Distanz

```python
predict_ds = predict_ds.sort_values('distanz')
```
